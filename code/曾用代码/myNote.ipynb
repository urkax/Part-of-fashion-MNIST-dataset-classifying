```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split

sns.set(style='white', context='notebook', palette='deep')
```

```python
#load data
train_list = pd.read_csv("data/train.csv")
Y_train = train_list["label"]
X_train = train_list.drop(labels = ["label"],axis = 1) 
del train_list
```

```python
g = sns.countplot(Y_train)
Y_train.value_counts()
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size = 0.1, random_state=4824)
g = sns.countplot(Y_train)
Y_train.value_counts()
```
